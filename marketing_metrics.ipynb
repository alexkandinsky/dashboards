- Показатель дневного Retention

```python
SELECT start_month,
       start_date,
       day_number,
       round(users::decimal / lag(users, day_number) OVER(PARTITION BY start_month,
                                                    start_date), 2) as retention
FROM(SELECT start_month,
            start_date,
            time::date - start_date as day_number,
            count(distinct user_id) as users FROM(SELECT min(month) as start_month,
                                                         min(time::date) as start_date,
                                                         user_id FROM(SELECT date_trunc('month', time::date)::date as month,
                                                                             time,
                                                                             user_id
                                                                       FROM user_actions) as t1
                                                  GROUP BY user_id) as t2
     join user_actions using(user_id)
GROUP BY start_month, start_date, day_number) as t3
ORDER BY start_month, start_date, day_number
```
